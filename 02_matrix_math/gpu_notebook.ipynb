````python
from cpu_matrix import cpu_matrix_multiply
from gpu_matrix import gpu_matrix_multiply
from gpu_matrix import matrix_gpu
import numpy as np
import time
````

### CPU Implementation:
* Dot product calculation using numpy's ```np.dot()``` method.
* Computation time over ```np.dot()``` only.

````python
A = np.random.rand(100,100)
B = np.random.rand(100,100)

cpu_result = cpu_matrix_multiply(A,B)
````

### GPU with ufunc (@vectorize):
* Uses Numba built in @vectorize decorator to create a ufunc.

````python
A_f32 = A.astype(np.float32)
B_f32 = B.astype(np.float32) 

start = time.time()
matrix_gpu(A_f32,B_f32)
end = time.time()
print(f'GPU time took {end-start}')
````
